### **HUMAN ACTIVITY RECOGNITION**

Data collection
Data has been collected as a part of experiment where 30 participants were involved performing six activities


*   STANDING
*   SITTING
*   LAYING
*   WALKING
*   WALKING_DOWNSTAIRS
*   WALKING_UPSTAIRS wearing a smartphone (Samsung Galaxy SII) on their
waist. Using its embedded accelerometer and gyroscope, 3-axial linear accelaration and 3-axial angular velocity was capture at a constant rate of 50Hz. Later, manual labeling was performed using the recorded video of experiments.

Calculating the euclidian norm from magnitude of corresponding 3-dimensional signals,aggregate features are obtained which are list as:

* tBodyAccMag

* tBodyAccJerkMag

* tBodyGyroJerkMag  

Also the data consists of other information about the above signals:



*   mean(): for Mean value
*   std(): for the Standard Deviation
*   mad(): for Median Absolute Deviation
*   max(): Maximum value
*   min(): Minimum value
*   sma(): Signal Magnitude Area
*   energy(): Enery measure
*   iqr(): Interquartile range
*   entropy(): Entropy measure
*   arCoeff(): AutoRegressive coefficient
*   correlation(): correlation coefficient between 2 signals
*   maxinds(): index of the largest magnitude of frequency component
*   meanFreq(): Mean Frequency
*   skewness(): skewness of the frequency domain signal
*   kurtosis(): kurtosis of frequency domain signal
*   bandsEnergy(): Energy of the frequency interval within the 64 bins of the FFT of each window.
*   angle(): angle between different signals

```python
# dictionary for activity and label
activity2labels = {'STANDING':1, 'SITTING':2, 'LAYING':3,
                   'WALKING':4, 'WALKING_DOWNSTAIRS':5, 'WALKING_UPSTAIRS':6}

# dictionary for label and activity
labels2activity = { v:k for k, v in activity2labels.items()}

# copy the train_df
train_df_copy = train_df.copy()

train_df_copy['Activity'] = train_df_copy['Activity'].apply(lambda x: activity2labels[x])
```

```python
from tqdm import tqdm

import numpy as np
from sklearn.manifold import TSNE

perplexities =  np.arange(5, 60, 5)
X_train = train_df_copy.iloc[:,:-2] # Not taking subject, and activity
Y_train = train_df_copy.iloc[:,-1]

def calculate_tsne(data=X_train, n_components=2, perplexity=perplexities, learning_rate=1000, random_seed=42):
    '''
    params:
        data: input high dimensional data
        n_components: number of dimension the input data to be reduced to.
        perplexity: parameter to control local and global structure
        learning_rate: paramter to control the step in optimization process
        random_seed: seed for random number generation
    returns:
        reduced: t-SNE embedding
    '''
    for i in tqdm(perplexity, desc="Calulating TSNE"):
        # initialize the t-SNE embedding
        tsne = TSNE(n_components=n_components, perplexity=i, learning_rate=learning_rate, random_state=random_seed)

        # fit and transform the model
        reduced = tsne.fit_transform(X_train)

    return reduced

def visualize_embedding(embedding, y):
    reduced_df = pd.DataFrame({'x':embedding[:,0], 'y':embedding[:,1] ,'label':y.apply(lambda x: labels2activity[x])})

    plt.figure(figsize=(15, 9))
    sns.lmplot(data=reduced_df, x='x', y='y', hue='label', fit_reg=False,
               palette="Set1", markers=['^', 'v', 's', 'o', '1', '2'])
    plt.title("t-SNE Visualization of Activities")
    plt.show()
```

```python
embedding = calculate_tsne()
```

```python
visualize_embedding(embedding, Y_train)
```

```python
from sklearn.svm import SVC

def federated_learning():
    num_clients = 3
    local_epochs = 5

    client_data = []
    for i in range(num_clients):
        X = X_train
        y = Y_train
        client_data.append((X, y))

    global_model = svc  # Initialize a global SVM model

    for epoch in range(local_epochs):
        client_models = []

        for client in range(num_clients):
            X, y = client_data[client]
            client_model = SVC(kernel='linear', C=1)  # Create a local SVM model for each client

            # Fit the local model to the client's data
            client_model.fit(X, y)
            client_models.append(client_model)

        # Merge support vectors and hyperplanes (global model remains the same)
        # You can choose different strategies to merge, but for linear SVM, there is no direct averaging
        # You can try combining support vectors and hyperplanes, but it's a more complex process
        # Here, we'll just keep the global model as-is

    # Make predictions using the global model
    global_predictions = global_model.predict(X_test)

    return global_model, global_predictions

global_model, predictions = federated_learning()
```

```python
def split_dataframe(df, n_splits):
    splits = np.array_split(df, n_splits)
    return splits
```

```python
def make_dataframe(split):
  df=pd.DataFrame(split,columns=train_df.columns)
  y=df['Activity']
  X=df.drop('Activity',axis=1)
  return X,y
```

```python
from sklearn.model_selection import train_test_split
def split_train_test(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
    return X_train, X_test, y_train, y_test
```

```python
def add_noise(X_test):
  laplace_scale = 0.1
  laplace_noise = np.random.laplace(loc=0, scale=laplace_scale, size=X_test.shape)
  noisy_data=X_test + laplace_noise
  return noisy_data
```

```python
from sklearn.svm import SVC

def model(X_train, Y_train):
    svm_classifier = SVC(kernel='linear', C=1, random_state=42)  # You can choose different kernel types and C values
    svm_classifier.fit(X_train, y_train)
    return svm_classifier
```

```python
def predict(model,X_test):
  y_pred = model.predict(X_test)
  return y_pred
```

```python
from sklearn.metrics import accuracy_score
def cal_accuracy(y_test,y_pred):
  accuracy = accuracy_score(y_test, y_pred)
  return accuracy
```

```python
num_splits = 5
result_splits = split_dataframe(train_df, num_splits)
```

```python
accuracy=[]
teacher_model=[]
for i in result_splits:
  X,y=make_dataframe(i)
  X_train, X_test, y_train, y_test=split_train_test(X,y)
  teacher=model(X_train,y_train)
  teacher_model.append(teacher)
  y_pred=predict(teacher,X_test)
  acc=cal_accuracy(y_test,y_pred)
  accuracy.append(acc)
```
